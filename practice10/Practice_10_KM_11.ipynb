### **Завдання 1 (1 бал)**

Продавець роздрібного товару влаштовує розпродаж. Для всіх товарів знижка однакова - 60%. Щоб вивісити нові цінники, продавець хоче скласти таблицю, в якій для кожного товару буде підрахована його нова вартість та сума знижки.

Написати програму, яка буде генерувати таку таблицю. Вартість одиниць товарів до розпродажу подано у списку products_list. 

1. Для підрахунку невідомих значень повинні використовуватись вбудовані функції вищих порядків.

2. Вартість з урахуванням знижки та суми знижок для кожного товару повинні зберігатись у двох окремих списках. Порядок товарів у нових списках повинен бути такий самий, як і в списку products_list.

3. Кожне підраховане числове значення має бути округлене до сотих.

4. Дані повинні виводитись у вигляді:

**Discount table:**

**29.25 11.7 17.55**

**48.99 19.6 29.39**

**...**


products_list = [29.25, 48.99, 99.98, 124.65, 214.30, 543.90, 799.85]


```
products_list = [29.25, 48.99, 99.98, 124.65, 214.30, 543.90, 799.85]

new_price = [*map(lambda a: [a, round(a*0.4, 2), round(a*0.6, 2)], products_list)]

for i in new_price:
    print(*i, sep=' ')
```

### **Завдання 2 (2 бали)**

Зазвичай рік має 365 днів. Однак у високосні роки додається ще один день - 29 лютого. Правила визначення високосного року такі:

- Будь-який рік, що ділиться на 400 - високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 100 - **не** високосний.

- З тих років, що залишились, будь-який рік, що ділиться на 4 - високосний.

- Усі інші роки, що залишились - **не** високосні.

1) Створіть функцію, в якій за допомогою вбудованих функцій вищих порядків будуть визначатися високосні роки серед перелічених років у списку years. Створена функція має повертати список високосних років.

2) Створіть функцію вищих порядків, яка прийматиме на вхід попередню функцію та визначатиме кількість днів у місяці. Також функція повинна приймати на вхід ще 2 параметри: ціле число від 1 до 12 (місяць) та рік, як ціле чотирицифрове число. Ці значення корисувач має вводити з клавіатури. Перевірку введених даних слід виконувати поза межами функцій.

```
import numpy as np

def get_leap(year:int) -> bool:
    return year % 400 == 0 or year % 100 != 0 and year % 4 == 0

def get_years(year:list) -> list:
    return [*filter(get_leap, year)]

def get_days(pred, month:int, year:int) :
    if month == 2: return 29 if pred(year) else 28
    elif month % 2 == 0: return 30
    else: return 31

def get_int(pred, str:str) -> int:
    while True:
        a = input(str + ': ')
        try:
            a = int(a)
        except:
            print('You wrot excepted number, try again')
            continue
        if pred(a) : return a
        else: 
            print('You wrot inappropriate value, try again')
            continue

years = np.arange(1900, 2020+1, 1)

print(f'Leap years: ')
print(*get_years(years), sep=', ')

year = get_int(lambda a: 0 < len(str(a)) <= 4, 'Input year')
month = get_int(lambda a: 1 <= a <= 12, 'Input month')

print(f'Days at {month} : {year} => {get_days(get_leap, month, year)}')
```
